```python
import pandas as pd
```

# Extract and Clean Startup Data

```python
path = 'Datasets/02-Investment_Dataset.csv'
```

```python
df = pd.read_csv(path, encoding = 'unicode_escape')
df.head()
```

```python
df.columns
```

```python
# Dropping Unwanted Columns

dropped_list = ['round_A', 'round_B',
       'round_C', 'round_D', 'round_E', 'round_F', 'round_G', 'round_H']
df = df.drop(dropped_list, axis=1)
df.head()
```

```python
df.columns
```

```python
# Dropping Unwanted Columns

dropped_list2 = ['permalink', 'homepage_url', 'undisclosed', 'convertible_note', 'debt_financing', 'angel', 'grant',
       'private_equity', 'post_ipo_equity', 'post_ipo_debt',
       'secondary_market', 'product_crowdfunding']
df = df.drop(dropped_list2, axis=1)
df.head()
```

```python
df.columns
```

```python
# Dropping Unwanted Columns

lst = ['founded_at', 'founded_month', 'founded_quarter', 'founded_year',
       'first_funding_at', 'last_funding_at', 'seed', 'venture',
       'equity_crowdfunding']
df = df.drop(lst, axis=1)
```

```python
df.head()
```

```python
# Cleaning Columns

df['category_list'] = df['category_list'].astype('str')
df['category_list'] = df['category_list'].replace('|', ' ')
df.dtypes
```

```python
# Cleaning Columns

lst = df['category_list'].tolist()
dirt = lst[0][0]
clean = []
for i in range(len(lst)):
    cleaned = lst[i].replace(dirt, ' ')
    clean.append(cleaned)
```

```python
# Dropping Unwanted Columns and Cleanung

df['clean_category_list'] = clean

df = df.drop(['category_list'], axis=1)
         
df.columns
```

```python
# Renaming and filtering data

df.rename(columns={'clean_category_list': 'category_list', ' funding_total_usd ': 'funding_total_usd',  ' market ': 'market'}, inplace=True)
df = df[['name', 'category_list', 'market', 'funding_total_usd', 'status', 'country_code', 'state_code', 'region', 'city', 'funding_rounds']]
df.head()
```

```python
# Outputting df to a csv

output_path = 'Datasets/03-Investment_Dataset_Cleaned.csv'
df.to_csv(output_path)
```
